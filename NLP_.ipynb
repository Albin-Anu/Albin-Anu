```python
#process 1
import pandas as pd

data = {
    "text": [
        "I love cricket",
        "Pullshot is difficult",
        "Cover Drive is my favourite",
        "I hate Wickets",
        "The Bowler is okay"
    ]
}

df = pd.DataFrame(data)
df
```

I love cricket - Positive
Pullshot is difficult - Negative
Cover Drive is my favourite - Positive
I hate Wickets - Negative
The Bowler is okay - Positive

```python
#Process 2
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

lambda - one time use function

```python
#process 3
!pip install afinn
from afinn import Afinn

af = Afinn()

sentences = ["I love cricket", "I hate wickets", "Bowler is okay"]
for s in sentences:
    print(s, af.score(s))
```

```python
#process 4
!pip install textblob
from textblob import TextBlob

sentences = ["I love cricket", "I hate wickets", "Bowler is okay"]
for s in sentences:
    blob = TextBlob(s)
    print(s, blob.sentiment)
```

Vader - Tuned for sentimental analysis provides a score between -1 and +1 where -1 most negative and +1 most positive.

Afinn - Provide a single score for the sentence by summing up the score of each words the scores are usually between -5 and +5

Textblob - It Provides 2 properties polarity and subjectivity
polarity range between -1 and 1
subjectivity range between 0 and 1

```python
#Let us compare !
from textblob import TextBlob

df['textblob_score'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_sentiment'] = df['textblob_score'].apply(
    lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral')
)
df
```

```python
import pandas as pd

data = {
    "text": [
        "I LOVE this film!!! ",
    "This course is SO boring...  ",
    "I think AI is amazing ",
    "Not sure if I like this. ",
    "Great work!!! Keep it up "
    ]
}

df = pd.DataFrame(data)
df
```

```python
df['clean_text'] = df['text'].str.lower()
df['clean_text'] = df['clean_text'].str.strip()
df[['text','clean_text']]
```

to treat every word as same we clean the sentences

```python
import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]
```

these punctuations doesnt provide necessary information so to remove this unneccesary contents we remove it

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]
```

don't carry much significant meaning on their own and removing them helps to reduce the size of the vocabulary and focus on the more important words that contribute to the overall meaning or sentiment of the text

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]
```
